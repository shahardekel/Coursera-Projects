```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT * 
FROM master
"""
master=pd.read_sql_query(QUERY,conn)
master['ConvertedComp'].plot(kind='hist', bins=30, color='blue', edgecolor='black')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

#plt.boxplot(master['Age'])

#master['Age'].value_counts()

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame 'master' is already loaded
# Create a box plot for the 'Age' column
plt.figure(figsize=(10, 6))
plt.boxplot(master['Age'].dropna())  # Drop NA values if there are any
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.xticks([1], ['Age'])  # Custom x-axis labels
plt.grid(axis='y')  # Optional: Add grid for better readability
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.scatter(master['Age'], master['WorkWeekHrs'], alpha=0.5)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame 'master' is already loaded
# Create a bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(master['WorkWeekHrs'], master['CodeRevHrs'], s=master['Age']*10, alpha=0.5, edgecolors="w", linewidth=0.5)
plt.title('Bubble Plot of Work Week Hours vs. Code Review Hours')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid(True)  # Optional: Add grid for better readability
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT * 
FROM DatabaseDesireNextYear
"""
desire= pd.read_sql_query(QUERY,conn)
# Group by 'DatabaseDesireNextYear' and count the number of respondents
database_counts = desire.groupby('DatabaseDesireNextYear')['Respondent'].count().reset_index()
top_databases = database_counts.sort_values(by='Respondent', ascending=False).head(5)

plt.figure(figsize=(8, 8))
plt.pie(top_databases['Respondent'], 
        labels=top_databases['DatabaseDesireNextYear'], 
        autopct='%1.1f%%', 
        startangle=140)

plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'master' DataFrame is already loaded
# Filter the DataFrame for the age group 30 to 35
age_filtered = master[(master['Age'] >= 30) & (master['Age'] <= 35)]

# Calculate the median WorkWeekHrs and CodeRevHrs
median_workweek = age_filtered['WorkWeekHrs'].median()
median_coderev = age_filtered['CodeRevHrs'].median()

# Prepare the data for the stacked chart
data = {
    'WorkWeekHrs': median_workweek,
    'CodeRevHrs': median_coderev
}

# Create a DataFrame for plotting
df = pd.DataFrame(data, index=[0])

# Plotting the stacked bar chart
df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.ylabel('Hours')
plt.xlabel('Age Group 30-35')
plt.xticks(rotation=0)
plt.legend(title='Hours Type', loc='upper left', labels=['WorkWeekHrs', 'CodeRevHrs'])
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'master' DataFrame is already loaded
# Filter the DataFrame for ages from 45 to 60
age_filtered = master[(master['Age'] >= 45) & (master['Age'] <= 60)]

# Calculate the median ConvertedComp for each age
median_converted_comp_by_age = age_filtered.groupby('Age')['ConvertedComp'].median()

# Plotting the median ConvertedComp as a line chart
plt.figure(figsize=(10, 5))
plt.plot(median_converted_comp_by_age.index, median_converted_comp_by_age.values, marker='o', linestyle='-')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.grid()
plt.xticks(median_converted_comp_by_age.index)  # Ensure all age values are shown on x-axis
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'master' DataFrame is already loaded
# Count the occurrences of each unique value in the MainBranch column
main_branch_counts = master['MainBranch'].value_counts()

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of Main Branches')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.grid(axis='x')  # Adding grid lines for better readability
plt.show()

```
